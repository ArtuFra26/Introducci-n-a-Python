## **Lectura de archivos de texto**

La apertura de un archivo de texto comienza por la definición de un **directorio** en el que se encuentra almacenado el archivo que necesitamos. Un directorio es una cadena de texto que hace referencia a sitio de almacenamiento en la memoria externa de la computadora. Cada directorio se encuentra formado por tres componentes: 

1. **Folder Path**: es la carpeta en la que se encuentra ubicado el archivo que se busca abrir
2. **File Name**: es el nombre del archivo de texto que se abrirá 
3. **Extension**: es la extensión indiciativo del tipo de archivo que se abrirá

El siguiente bloque de código pone en práctica la verificación de la existencia de un directorio en la memoria de la computadora:

```python
#Importar bibliotecas requeridas
import os.path

#Definir un directorio en la computadora local
path = "D:/Documentos/Otros/Aplicaciones/DESI"

#Verificar la existencia del directorio
isdir = os.path.isdir(path)
print(isdir)
```

El valor obtenido luego de la ejecución del código, `True`, es indicativo de que, luego de hacer la búsqueda en la memoria de la computadora, se encontró una carpeta con el nombre indicado.

Para practicar la lectura de archivos de texto, utilizaremos dos archivos con extensiones distintas: `.txt` y `.csv`. Los datos de organizaciones son almacenados en este tipo de archivos la mayor parte del tiempo.

## ¿Qué son los archivos de estructura tabular?

Los archivos de estructura tabular son el tipo de documento más común para el trabajo con datos. Este tipo de archivo es un variedad de **archivo de texto** y su representación más habitual es la de la hoja de cálculo utilizada en Excel. 

Los archivos de texto de estructura tabular son, la mayor parte del tiempo, almacenados en archivos de extensión `.csv`. Su manipulación requiere saber cómo se encuentra organizada su **estructura**. 

Una tabla o matriz de datos se conforma de tres elementos:

1. **Filas**: hacen referencia a las *unidades* de la base de datos, cada fila representa un registro individual. Definen el nivel del análisis. 
2. **Columnas**: indican las variables o características (*features*) que describen a las unidades de la base. Definen la extensión o dimensionalidad del análisis.
3. **Celdas**: contiene el valor de una variable para una unidad o registo individual

## Operadores

Los **operadores** son símbolos especiales que permiten ejecutar operaciones que toman como inusmos a **variables** o valores. Existen varias clases de operadores que se diferencían por el tipo de tareas que pueden ejecutar.

### Operadores aritméticos

Los operadores aritméticos son utilizados para ejecutar **operaciones matemáticas** en Python. 

```python
#Definir variables numéricas
a = 8
b = 10

#Adición
print("Suma:", a + b)

#Sustracción
print("Resta:", a - b)

#Multiplicación
print("Multiplicación:", a * b)

#División
print("División:", a / b)

#Exponenciación
print("Exponenciación:", a**b)
```

### Operadores de asignación

Los operadores de asignación son utilizados para asignar **valores** a **variables**. 

```python
#Asignación regular
c = 5

#Asignación con suma. 
c += a # Es esquivalente a 'c = c + 8'
print("Asignación con suma:", c)

#Asignación con resta. 
c -= a # Es equivalente a 'c = c - 8'
print("Asignación con resta:", c)

#Asignación con multiplicación
c *= a # Es equivalente a 'c = c*8'
print("Asignación con multiplicación:", c)

#Asignación con división
c /= a # Es equivalente a 'c = c/8'
print("Asignación con división", c)

#Asignación con exponenciación
c **= a # Es equivalente a 'c = c**8'
print("Asignación con exponenciación:", c)
```

### Operadores de comparación

Los operadores de comparación comparan **dos valores o variables**. Son utilizados para hacer **pruebas condicionales**.

El output de la ejecución de una prueba condicional es una variable de **tipo booleano** que cuenta con dos valores posibles: `True` (si se cumple la condición) o `False` (si no se cumple la condición).

```python
#Verificar si 'a' es igual que 'c' (igualdad estricta)
print("Igualdad estricta:", a == c)

#Verificar si 'a' es diferente que 'c'
print("Desigualdad:", a != c)

#Verificar si 'a' es mayor que 'c'
print("Mayor que:", a > c)

#Verificar si 'a' es menor que 'c'
print("Menor que:", a < c)

#Verificar si 'd' es mayor o igual que 'c'
d = 5
print("Mayor o igual que:", d >= c)

#Verificar si 'e' es menor o igual que c
e = 4
print("Menor o igual que:", e <= c)
```

### Operadores lógicos

Estos operadores son utilizados para **verificar** si múltiples condiciones se cumplen o no. Permiten hacer **pruebas condicionales compuestas**.

Su valor de salida también es una variable de tipo booleano.

```python
#Verificar si se cumplen dos condiciones
print("Conjunción:", (a > 2) and (b < 6))

#Verificar si se cumple una de las dos condiciones
print("Disyunción:", (a > 2) or (b < 6))

#Verificar si NO se cumple la condición
print("Negación:", not (a > 2))
```

### Operadores especiales

Existen operadores que permiten ejecutar operaciones de comparación de datos más detalladas. Su valor de salida también es una variable de tipo booleano.

```python
#Operadores de identidad
#Se utilizan para verificar si dos valores son 'iguales' e 'idénticos'

#Verificar si dos valores son iguales e idénticos
f = "5"
print("Iguales e idénticos:", c is f)
print("No iguales ni idénticos:", c is not f)

#Operadores de membresía
#Se utilizan para verificar si un valor o variable pertenece a una estructura de datos

#Definir una lista de valores
ls = [5, 8, 10]
print("Pertenece 'c' a la lista:", c in ls)
print("No pertenece 'c' a la lista:", c not in ls)
```
