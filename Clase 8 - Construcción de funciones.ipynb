## Pregunta 3: Generar numeralia

**Instrucciones:** Dada la columna seleccionada por el usuario, amplíe la función generada para generar una numeralia  de resumen (basada en el dataset original) que incluya la cantidad de observaciones y la edad promedio por grupo. La variable de agrupación corresponde a la seleccionada por el usuario a partir del menú desplegado. Ordene los valores de conteo en orden descendente.

Utilice las funciones: `df.groupby, df.sort_values, df.columns, round; df.iloc; df.mean`

```python
#Ampliar la función para obtener numeralia de resumen
def col_usr():
    #Imprimir dataframe de opciones
    print(df_names)
    #Solicitar el input del ususario
    col = int(input("¿Qué columna desea utilizar? "))
    #Definir dataframe con columnas de resumen
    df_gr = df.iloc[:, [4, col]]
    #Agrupar columna de edad dada columna elegida y generar conteo
    df_cnt = df_gr.groupby(df_gr.columns[1]).count()
    #Ordenar valores de conteo en order descendente
    df_cnt = df_cnt.sort_values(df_cnt.columns[0], ascending = False)
    #Agrupar columna de edad dada columna elegida y estimar media
    df_cnt["Promedio"] = round(df_gr.groupby(df_gr.columns[1]).mean(), 2)
    #Retornar valores finales
    return df_cnt
    
#Ejecutar función
col_usr()
```

## Pregunta 4: Condición

**Instrucciones:** Amplíe la función para verificar si la tabla de la numeralia contiene más de cinco filas. En caso de que se cumpla esta condición, tome las primeras cinco filas y genere un gráfico de barras utilizando los valores de conteo de observaciones de cada grupo. Apoyese en el siguiente recurso para generar la gráfica de barras: https://python-graph-gallery.com/barplot/

Utilice las funciones: `df.shape` y `plt.bar`.

```python
#Ampliar la función para verificar extensión del output 
def col_usr():
    #Imprimir dataframe de opciones
    print(df_names)
    #Solicitar el input del ususario
    col = int(input("¿Qué columna desea utilizar? "))
    #Definir dataframe con columnas de resumen
    df_gr = df.iloc[:, [4, col]]
    #Agrupar columna de edad dada columna elegida y generar conteo
    df_cnt = df_gr.groupby(df_gr.columns[1]).count()
    #Ordenar valores de conteo en order descendente
    df_cnt = df_cnt.sort_values(df_cnt.columns[0], ascending = False)
    #Agrupar columna de edad dada columna elegida y estimar media
    df_cnt["Promedio"] = round(df_gr.groupby(df_gr.columns[1]).mean(), 2)
    
    #Verificar si el output final tiene más de cinco filas
    if df_cnt.shape[0] > 5:
        #Generar gráfico cuando se cumple la condición
        import matplotlib.pyplot as plt
        df_plt = plt.bar(df_cnt.index[0:5], df_cnt.iloc[0:5, 0])
        #Retornar valores finales
        return df_cnt, df_plt
    #Ejecución alternativa
    else:    
        #Retornar valores finales
        return df_cnt
    
#Ejecutar función
col_usr()
```
