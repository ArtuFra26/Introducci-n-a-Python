## Lectura de archivos de texto

Para practicar la lectura de archivos de texto en Python utilizaremos el primer capítulo de la novela *Contraband* de Clarence Budington Kelland (2023), disponible en la página de Project Gutenberg.

La apertura de un archivo de texto comienza por la definición de un **directorio** en el que se encuentra almacenado el archivo que necesitamos. Un directorio es una cadena de texto que hace referencia a sitio de almacenamiento en la memoria externa de la computadora. Cada directorio se encuentra formado por tres componentes: 

1. **Folder Path**: es la carpeta en la que se encuentra ubicado el archivo que se busca abrir
2. **File Name**: es el nombre del archivo de texto que se abrirá 
3. **Extension**: es la extensión indiciativo del tipo de archivo que se abrirá

El siguiente bloque de código pone en práctica la verificación de la existencia de un directorio en la memoria de la computadora:

```python
#Importar bibliotecas requeridas
import os.path

#Definir un directorio en la computadora local
path = "D:/Documentos/Otros/Aplicaciones/DESI"

#Verificar la existencia del directorio
isdir = os.path.isdir(path)
print(isdir)
```

El valor obtenido luego de la ejecución del código, `True`, es indicativo de que, luego de hacer la búsqueda en la memoria de la computadora, se encontró una carpeta con el nombre indicado.

Para practicar la lectura de archivos de texto, utilizaremos dos archivos con extensiones distintas: `.txt` y `.csv`. Los datos de organizaciones son almacenados en este tipo de archivos la mayor parte del tiempo.

La lectura de archivos de texto se puede ejecutar de diversas formas en dependencia del tipo de archivo y de lo que se desea hacer con el contenido del mismo: lectura, escritura o una combinación de ambas.

En el caso de un archivo de tipo `.txt`, el comando requerido para abrir el archivo en modo lectura es el siguiente:

```python
#Abrir archivo de texto simple
ctx = open("D:/Documentos/Otros/Aplicaciones/DESI/Contraband.txt", mode = "r", encoding = 'utf-8')
```

El comando anterior abre el archivo de texto en modo lectura (`mode = 'r'`), de manera que será posible ver el contenido del mismo pero no editarlo. Por otro lado, es posible observar que, como parte del comando, es posible especificar el encoding del archivo. En este caso, se especifica el tipo 
`utf-8` como el sistema de referencia del archivo.

A fin de explorar el contenido del archivo, es posible utilizar la siguiente instrucción:

```python
#Obtener información sobre el contenido del archivo
ctx.read()
```

Al utilizar el modo escritura (`mode = 'w'`), se sobreescribe en su totalidad el contenido del archivo original para generar un nuevo archivo editable. Por otro lado, este comando hace posible la creación de un nuevo archivo de texto en el directorio indicado. El comando requerido para llamar esta función es el siguiente:

```python
#Abrir archivo de texto simple
ctx_n = open("D:/Documentos/Otros/Aplicaciones/DESI/Contraband2.txt", mode = "w")
```

Al revisar la carpeta definida como directorio en el la línea de código, será posible encontrar un nuevo archivo de nombre Contraband2. En caso de haber definido el archivo original, Contraband.txt, como archivo de trabajo, el contenido del mismo habría sido eliminado para ser reemplazado con nuevo texto.

Es posible utilizar el siguiente comando para introducir texto en el archivo creado:

```python
#Escribir texto en un archivo
ctx_n.write('Este es un archivo de práctica')

#Cerrar el archivo de práctica
ctx_n.close()
```

Para visualizar los cambios en el archivo de práctica es necesario cerrar el documento en Python a través de la instrucción `.close()`. Hecho esto, al abrir el archivo `Contraband2.txt` será posible observar que éste contiene el texto deseado: 'Este es un archivo de práctica'.

Es posible combinar los modos de lectura y escritura para trabajar con archivos que ya cuentan con texto y en los que se desea incorporar nuevo contenido. Para ello es necesario utilizar el modo `append` and `read`, a través del siguiente comando:

```python
#Leer y añadir texto en un archivo 
with open("D:/Documentos/Otros/Aplicaciones/DESI/Contraband2.txt", mode = "a+") as ntx:
    ntx.write("Una pequeña adición")
    
#Leer el nuevo texto que se añadió al archivo de práctica
with open("D:/Documentos/Otros/Aplicaciones/DESI/Contraband2.txt") as ntx:
    print(ntx.read())
```

La salida anterior permite constatar que, al trabajar con el archivo de práctica previamente, ha sido posible incorporar el texto, 'Una pequeña adición', sin eliminar el anterior. Posteriormente, se abrió el archivo en modo lectura para visualizar todo el contenido del mismo.

En este caso, se incorporó la instrucción `with` a los comandos de lectura y escritura. Esta adición ofrece la ventaja de concatenar muchas instrucciones como la apertura del archivo, su edición o lectura y su cierre; aun si ésta última no se indica explícitamente.

## ¿Qué son los archivos de estructura tabular?

Los archivos de estructura tabular son el tipo de documento más común para el trabajo con datos. Este tipo de archivo es un variedad de **archivo de texto** y su representación más habitual es la de la hoja de cálculo utilizada en Excel. 

Los archivos de texto de estructura tabular son, la mayor parte del tiempo, almacenados en archivos de extensión `.csv`. Su manipulación requiere saber cómo se encuentra organizada su **estructura**. 

Una tabla o matriz de datos se conforma de tres elementos:

1. **Filas**: hacen referencia a las *unidades* de la base de datos, cada fila representa un registro individual. Definen el nivel del análisis. 
2. **Columnas**: indican las variables o características (*features*) que describen a las unidades de la base. Definen la extensión o dimensionalidad del análisis.
3. **Celdas**: contiene el valor de una variable para una unidad o registo individual

La estructura de una base de datos tabular corresponde a lo observado en la siguiente imagen:

```python
#Insertar una imagen
from IPython import display
display.Image("https://statisticsglobe.com/wp-content/uploads/2021/12/table-3-data-frame-append-rows-pandas-dataframe-loop-python.png")
```

El rango vertical que va de 0 a 4 corresponde al **índice** de filas y es indicativo de la posición de cada unidad en el conjunto de datos. Por otro lado, las celdas en azul integran los nombres de variable o **header**.
